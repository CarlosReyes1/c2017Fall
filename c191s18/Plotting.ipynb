```python
%matplotlib inline
# Important for using matplotlib w/ Ipython Notebook
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import graphviz as gv

import random
import numpy as np

import math

import sys
import os

from __future__ import division
```

```python
x = [i * math.pi*2/100 for i in range(100)]
print x[::10]
y = [math.sin(i) for i in x]
plt.plot(x,y)
plt.show()
```

```python
radius = 1
theta = np.linspace(0, 2*np.pi*radius, 1000)
print theta[::100]
plt.subplot(111, projection='polar')
plt.plot(theta, np.cos(7*theta), "g-")
plt.plot(theta, 0.85*np.cos(14*theta), "b-")
plt.plot(theta, 0.3*np.cos(14*theta), "r-")
plt.show()
```

```python
radius = 1
theta = np.array([t/1000*14*np.pi*radius for t in range(1000)])
print theta[::100]
plt.subplot(111, projection='polar')
plt.plot(theta, [1*i for i in theta], "g-")
plt.show()
```

```python
plt.rc('grid', color='#316931', linewidth=0, linestyle='-')
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=0)

radius = 1
theta = np.array([t/1000*4*np.pi*radius for t in range(1000)])
print theta[::100]
plt.subplot(111, projection='polar')
plt.plot(theta, [math.e**(0.30635*i) for i in theta], color="gold")
plt.title("Golden Spiral")
plt.show()
```

```python
plt.rc('grid', color='#316931', linewidth=0, linestyle='-')
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=0)

radius = 1
theta = np.array([t/1000*2*np.pi*radius for t in range(1000)])
print theta[::100]
ax = plt.subplot(111, projection='polar')
for j in range(12):
    plt.plot([i+j*2*math.pi/12 for i in theta], [math.sin(t*i) for i in theta], color=["red", "green", "blue", "gold"][j % 4])
plt.title("Spirograph")
ax.axis('off')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

plt.rc('ytick', labelsize=12)

size = 100
r=100
Xnp = r*np.random.rand(size,1)
print Xnp[:10]
Xdf = pd.DataFrame(Xnp)
Xdf[1] = StandardScaler().fit_transform(Xdf)
X0 = Xdf[0].values.reshape(size,1)
y = 4 + 3 * X0+ np.random.randn(size, 1)*r
print Xdf[1][:10]
print y[:10]

fig = plt.figure(figsize=(10,5))
fig.add_subplot(121)
col=1
lin_reg = LinearRegression()
lin_reg.fit(Xdf[col].values.reshape(size,1),y)
print lin_reg.intercept_, lin_reg.coef_
pred = lin_reg.predict(Xdf[col].values.reshape(size,1))
plt.plot(Xdf[col], pred, "r-")
plt.plot(Xdf[col], y, "b.")
#plt.axis([0, 2, 0, 15])
fig.add_subplot(122)
#plt.figure(figsize=(5,5))
plt.plot(Xdf[0], pred, "r-")
plt.plot(Xdf[0], y, "b.")
plt.show()
```
