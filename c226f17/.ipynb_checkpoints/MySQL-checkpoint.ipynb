# Images in MySQL

```python
def GraphImage(V, E, w={}, D=True, C={}, N=[]):
    'Plots a graph'
    G = gv.Digraph(filename='graph', format='jpg') if D else gv.Graph(filename='graph', format='jpg')
        
    for v in V:
        if v in N:
            G.node(str(v), color='blue', fontcolor='blue') 
        else:
            G.node(str(v))
    for u in E:
        for v in E.get(u, []):
            if not D and u>v: continue
            if (u,v) in C or ((v,u) in C and not D):
                G.edge(str(u),str(v), label = str(w.get((u,v), '')), color='blue')
            else:
                G.edge(str(u),str(v), label = str(w.get((u,v), '')))
    G.render()
    img=mpimg.imread('graph.jpg')
    return img

def RandomEdgePermute(N, p, selfE=False, D=True):
    V = range(1,N+1)
    Elist = [(u,v) for u in range(1,N+1) for v in range(1,N+1) if (u!=v or selfE) and (u<v or D)]
    random.shuffle(Elist)
    E = {}
    for i in range(int(p*len(Elist))):
        u, v = Elist[i]
        e = E.get(u,[])
        e.append(v)
        e.sort()
        E[u] = e
        if not D:
            e = E.get(v,[])
            e.append(u)
            e.sort()
            E[v] = e
    return V, E
```

```python
InsertR="insert into Images (sid, image) value (%s, %s)"

conn = mysql.connector.connect (host = "localhost",
                                user = "root", passwd = "cs126", db="graphs")
cursor = conn.cursor()
for i in range(11):
    V, E = RandomEdgePermute(20, 0.1, True)
    img = GraphImage(V, E)
    out=img.dumps()
    cursor.execute(InsertR, (i+1, out))
conn.commit()
conn.close()

conn = mysql.connector.connect (host = "localhost",
                                user = "root", passwd = "cs126", db="graphs", connection_timeout=28800)
cursor = conn.cursor()
img=mpimg.imread('dog.jpg')
output = img.dumps()
cursor.execute(InsertR, (13, output) )
cursor.execute(QueryList)
data = pd.DataFrame(cursor.fetchall(), columns=['KEY'])
print data
conn.commit()
conn.close()
```

```python
def Path(P, u):
    ret = [u]
    v = P[u]
    while v:
        ret.append(v)
        u, v = v, P[v]
    ret.reverse()
    return ret

def BFS(V, E, s):
    Color = {}
    Distance = {}
    Parent = {}
    for u in V:
        Color[u] = 0
        Distance[u] = sys.maxint
        Parent[u] = 0
    Color[s] = 1
    Distance[s] = 0
    Q = [s]
    while Q:
        u =Q.pop()
        for v in E.get(u,[]):
            if not Color[v]:
                Color[v] = 1
                Distance[v] = Distance[u] + 1
                Parent[v] = u
                Q.insert(0, v)
        Color[u] = 2
    return Color, Distance, Parent


V, E = RandomEdgePermute(20, 0.1, True)
C, D, P = BFS(V, E, 1)
P = Path(P, 13)
pEdges = [(P[i], P[i+1]) for i in range(len(P)-1)]
print pEdges
```

```python
img = GraphImage(V, E, C=pEdges, N=[1,13])
fig, ax1 = plt.subplots()
fig.set_size_inches(13,13)
imgplot = plt.imshow(img)
ax1.axis('off')
plt.show() 
```
